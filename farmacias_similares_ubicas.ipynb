```python
import pandas as pd
import requests
import folium
```

```python
url_ubicas="https://www.farmaciasdesimilares.com/getpickuppoints"
```

```python
#Hacer la peticion
response = requests.get(url_ubicas)
data = response.json()
```

```python
nombre_farmacia=[]
ciudad=[]
estado=[]
colonia=[]
calle=[]
cp=[]
numero=[]
latitud=[]
longitud=[]

for farmacias in range(len(data)):
    nombre_farmacia.append(data[farmacias]["name"])
    ciudad.append(data[farmacias]["address"]["city"])
    estado.append(data[farmacias]["address"]["state"])
    colonia.append(data[farmacias]["address"]["neighborhood"])
    calle.append(data[farmacias]["address"]["city"])
    numero.append(data[farmacias]["address"]["number"])
    cp.append(data[farmacias]["address"]["postalCode"])
    latitud.append(data[farmacias]["address"]["location"]["latitude"])
    longitud.append(data[farmacias]["address"]["location"]["longitude"])
```

```python
simi_data = pd.DataFrame({
    "nombre_farmacia": nombre_farmacia,
    "calle":calle,
    "numero":numero,
    "colonia":colonia,
    "ciudad": ciudad,
    "codigo_postal":cp,
    "latitud": latitud,
    "longitud": longitud

})
simi_data
```

```python
url_carita_simi="https://storage.googleapis.com/www-saludiario-com/wp-content/uploads/2022/11/47207d84-logo-simi-e1667606405832-700x700.png"
```

```python
#Mapa de estadios con folium utilizando la url de google hybrid
mapa=folium.Map(location=[19.432608, -99.133209],zoom_start=6,tiles="https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}",attr="Google")
for i in range(0,len(simi_data)):
    #Marcador con logo del estadio
    folium.Marker(location=[simi_data["latitud"][i],simi_data["longitud"][i]],popup=folium.Popup(f"<b>Farmcia:</b> {simi_data['nombre_farmacia'][i]}<br><b>Calle:</b> {simi_data['calle'][i]}<br><b>Colonia:</b> {simi_data['colonia'][i]}"),
                    icon=folium.features.CustomIcon(url_carita_simi,icon_size=(40,40))).add_to(mapa)
mapa
```

```python
mapa.save("farmacias_simi.html")
```
